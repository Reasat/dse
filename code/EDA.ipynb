```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import datetime as dt
```

```python
df=pd.read_excel(os.path.join('..','data.xlsx'))
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns=['Date', 'Total_Trade', 'Total_Volume',
       'Total_Value_in_Taka(mn)',
       'Total_Market_Cap._Taka(mn)', 'DSEX_Index',
       'DSES_Index', 'DS30_Index', 'DGEN_Index']
```

```python
usable_dsex_index=df.index[pd.notnull(df['DSEX_Index'])]
print('No of usable DSEX Index: ', len(usable_dsex_index))
usable_dses_index=df.index[pd.notnull(df['DSES_Index'])]
print('No of usable DSES Index: ', len(usable_dses_index))
usable_ds30_index=df.index[pd.notnull(df['DS30_Index'])]
print('No of usable DS30 Index: ', len(usable_ds30_index))
usable_dgen_index=df.index[pd.notnull(df['DGEN_Index'])]
print('No of usable DGEN Index: ', len(usable_dgen_index))
```

```python
n_dsex=len(usable_dsex_index)
date=[dt.datetime.strptime(d,'%d-%m-%Y').date() for d in list(df['Date'])[:int(n_dsex/5)]]
dsex=list(df['DSEX_Index'])[:int(n_dsex/5)]
plt.figure(figsize=(20,4))
plt.plot(date,dsex,label='DSEX_Index')
plt.xticks(date,rotation='vertical')
plt.gca().xaxis.set_major_formatter(mdate.DateFormatter('%d/%m/%Y'))
plt.gca().xaxis.set_major_locator(mdate.DayLocator(interval=3))
plt.grid(True)
plt.legend()
plt.show()
```

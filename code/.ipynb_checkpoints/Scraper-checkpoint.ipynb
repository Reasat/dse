```python
from bs4 import BeautifulSoup
import urllib.request 
import re
```

```python
wiki = "http://dsecsv.simdif.com/index.html"
page = urllib.request.urlopen(wiki)
soup = BeautifulSoup(page,'html.parser')
```

```python
data = soup.find_all("div", { "class" : 'block_inner clearfix'})
```

```python
links=[(link.get('href'),link.get_text()) for div in data for link in div.findAll('a')]  
```

```python
links
```

```python
dl_dir='..\\data\\2017\\'
for link,fl_name in links:
    if 'DSE'in fl_name:       
        urllib.request.urlretrieve(link,dl_dir+fl_name+'.csv')
    
```
